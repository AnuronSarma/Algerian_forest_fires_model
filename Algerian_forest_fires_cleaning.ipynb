```python
#Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Read the csv file
df = pd.read_csv("/content/Algerian_forest_fires_dataset_UPDATE.csv", header=1)
df.head()
```

```python
df.info()
```

```python
#Data cleaning
df[df.isnull().any(axis=1)]
```

```python
#The dataset will be converted into sets as two sets of data are present in it.
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
df.info()
```

```python
df[["Region"]]=df[["Region"]].astype(int)
```

```python
df.isnull().sum()
```

```python
#Removing the null values
df=df.dropna().reset_index(drop=True)
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.head()
```

```python
df.columns
```

```python
#Fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
#Changing the required columns to integer datatype
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
#Changing the required columns to floating data type
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
#Iterating through the objects and skipping classes column and changing rest to float
for i in objects:
  if i!='Classes':
    df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
#Saving the entire dataset as a cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv')
```
